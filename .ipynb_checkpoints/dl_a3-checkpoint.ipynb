```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import roc_curve, auc, confusion_matrix, ConfusionMatrixDisplay
from keras.models import Sequential
from keras.layers import Input, InputLayer, Conv2D, MaxPool2D, Flatten, Dense, BatchNormalization, Dropout
from keras.utils import load_img, img_to_array,to_categorical
from keras.initializers import HeNormal
import cv2
from skimage import exposure
from skimage.filters import gaussian
from sklearn.metrics import accuracy_score, recall_score
from keras.optimizers import Adam
from keras.callbacks import EarlyStopping
import warnings

warnings.filterwarnings('ignore')
```

# Reading and storing data

```python
data_dir = "Val"
folders = ["COVID-19", "Non-COVID", "Normal"]
```

### Preprocessing Images, We don't need to apply preprocessing on Mask images since they are binary and represent only the lungs.

```python
def PreprocessImage(img):
    #Resizing Imags
    size = (256,256)
    img = cv2.resize(img,size) 
    
    #Converting Image to Grayscale
    img = img.astype(np.uint8)
    if len(img.shape) == 3 and img.shape[2] == 3:
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    else:
        gray = img
    
    #Normalizing the pixel values
    img = gray / 255
    
    #Thresholding the values
    #We're using Otsu Thresholding to automatically detect the optimal threshold value to seperate foregroung 
    #and background pixels.
    _, img = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    
    #Applying Gaussian Blur
    img = gaussian(img, sigma=1, multichannel=False)
    
    #implementing Histogram Equalization
    img = (exposure.equalize_hist(img) * 255).astype(np.uint8)
    return img
```

```python
# reading images from all folders and storing them as numpy arrays in a list

images = []
lung_masks = []
condition = []

for f in folders:
    images_dir_path = os.path.join(data_dir, f, "images")
    masks_dir_path = os.path.join(data_dir, f, "lung masks")
    
    images_list = os.listdir(images_dir_path)

    for i in images_list:
        img = load_img(os.path.join(images_dir_path, i))
        img_arr = img_to_array(img).astype(int)

        mask = load_img(os.path.join(masks_dir_path, i))
        mask_arr = img_to_array(mask).astype(int)
        
        img_arr = PreprocessImage(img_arr)
        images.append(img_arr)
        
        if len(mask_arr.shape) == 3 and mask_arr.shape[2] == 3:
            mask_arr = cv2.cvtColor(mask_arr.astype(np.uint8), cv2.COLOR_BGR2GRAY)
        lung_masks.append(mask_arr)
        
        condition.append(f)
```

```python
print(len(images))
print(len(lung_masks))
print(len(condition))
```

```python
# convert lists to numpy arrays

images_data = np.array(images)
lung_masks_data = np.array(lung_masks)

print(images_data.shape)
print(lung_masks_data.shape)
```

```python
# one hot encoding the condition(i.e. target variable)

condition_data = pd.get_dummies(condition, dtype=int)
condition_data.head()
```

```python
condition_columns = condition_data.columns
```

```python
condition_data = condition_data.to_numpy()
condition_data.shape
```

```python
plt.imshow(images_data[0])
```

```python
plt.imshow(lung_masks_data[0])
```

```python
np.amax(lung_masks_data[0])
```

```python
# converted the white values(i.e. 255) to 1
lung_masks_data[lung_masks_data==255] = 1
```

```python
np.amax(lung_masks_data[0])
```

```python
# applied the mask to xray images so that only lungs remain in the image and the surrounding values are set to 0

input_data = images_data * lung_masks_data
input_data.shape
```

```python
plt.imshow(input_data[0])
```

# Train Test Split

```python
X = input_data
Y = condition_data
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, shuffle=True, stratify=Y)

print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
input_data[0].shape
```

```python
condition_data[0].shape
```

# CNN model

```python
outputs = 3
batch_size = 128
epochs = 10
```

```python
initializer1 = HeNormal()
initializer2 = HeNormal()
```

```python
X[0].shape
```

```python
inputShape = (256, 256, 1)

model = Sequential()

model.add(InputLayer(input_shape = inputShape))

model.add(Conv2D(32,
                 kernel_size=(2,2),
                 strides=(1,1),
                 padding='same',
                 activation='relu'
                 ))

model.add(Conv2D(64,
                 kernel_size=(3,3),
                 strides=(2,2),
                 padding='same',
                 activation='relu'
                 ))
model.add(MaxPool2D(pool_size=(3,3), padding='same'))
model.add(BatchNormalization())
model.add(Dropout(0.2))

model.add(Conv2D(128,
                 kernel_size=(3,3),
                 strides=(3,3),
                 padding='same',
                 activation='relu'
                 ))
model.add(MaxPool2D(pool_size=(3,3), padding='same'))
model.add(BatchNormalization())
model.add(Dropout(0.2))

model.add(Flatten())
model.add(Dense(128, activation='relu', kernel_initializer=initializer1))
model.add(Dense(64, activation='relu', kernel_initializer=initializer2))

model.add(Dense(outputs, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, 
          batch_size=batch_size, 
          epochs=epochs, 
          validation_data=(X_test, Y_test))
```

```python
Y_pred = model.predict(X_test)
Y_pred
```

```python
cols = condition_columns.to_list()
```

# AUC-ROC Curve

```python
fig, ax = plt.subplots(1,3)

fig.set_figwidth(16)

for i in range(len(cols)):
    fpr, tpr, thresholds = roc_curve(Y_test[:, i], Y_pred[:,i])
    auc_v = auc(tpr, fpr)

    ax[i].plot(fpr, tpr, label='Area = {:.3f}'.format(auc_v))
    ax[i].set_xlabel('False positive rate')
    ax[i].set_ylabel('True positive rate')
    ax[i].set_title(cols[i])
    ax[i].legend(loc='best')

plt.show()
```

# Confusion Matrix

```python
cm = confusion_matrix(Y_pred.argmax(axis=1), Y_test.argmax(axis=1))

ConfusionMatrixDisplay(cm, display_labels=cols).plot()
plt.show()
```

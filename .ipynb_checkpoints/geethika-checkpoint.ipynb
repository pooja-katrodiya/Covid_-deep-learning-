# Data Preprocessing

## Importing the libraries

```python
import pandas as pd
import numpy as np
import os
import cv2
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

## Loading the data

### Setting up the paths of the files

```python
# set up of the images path
root_path = '/Users/hemanthalaparthi/Documents/Courses/S3/DeepLearning/Ungraded_Assign2/branddataset'
images_path = os.path.join(root_path, 'images/')
```

### Loading the csv file

```python
brand_info = pd.read_csv(os.path.join(root_path, 'brand_info.csv'))
```

```python
brand_info.head()
```

The images whose ID is within this csv file must be considered as original products. The remaining ones are fakes.

```python
# creating an array with the IDs of the original products
id_original = np.array(brand_info['ID'])
```

```python
# creating an array with the IDs of the fake products
id_fake = np.array(())

for item in os.listdir(images_path):  # looping through each file in this path
    if int(item.split('.')[0]) not in id_original:
        id_fake = np.append(id_fake, int(item.split('.')[0]))
```

```python
# checking the number of images in the folder
print('Total number of images = ', len(os.listdir(images_path)))
print('Number of original images = ', len(id_original))
print('Number of fake images = ', len(id_fake))
# print(f'{len(id_original) / len(os.listdir(images_path)) * 100:.1f}% of the images are original.')
```

```python
# visualizing the images distribution
bars = plt.bar(['Original', 'Fake'], [len(id_original) / len(os.listdir(images_path)) * 100, len(id_fake) / len(os.listdir(images_path)) * 100], color=['blue', 'red'])
plt.title('Distribution of image classes')
plt.ylabel('Proportion [%]')

for item in bars:
    height = item.get_height()
    plt.text(item.get_x() + item.get_width() / 2.0, height, f'{height:.1f}%', ha='center', va='bottom')

plt.show()
```

Only 34.1% of the images are considered as original.

### Loading the images

```python
# checking the image shape
img_test = cv2.imread(images_path + '9733.jpg', cv2.IMREAD_GRAYSCALE)
print('Image shape =', img_test.shape)
```

It is expected that the shape of all the images are 80 x 60. However, some of them might have a different size, like the example below.

```python
# checking the image shape before padding
img_test = cv2.imread(images_path + '1801.jpg', cv2.IMREAD_GRAYSCALE)
print('Image shape =', img_test.shape)
plt.imshow(img_test, cmap='gray')
```

To solve this problem, we will pad with the value of 255 (white) on the dimension that requires it.
References: 
- https://medium.com/@Orca_Thunder/image-padding-techniques-constant-padding-part-4-7906f12f2fab
- https://stackoverflow.com/questions/51650325/how-to-pad-an-image-using-np-pad-in-python

```python
# padding with 255
img_test = np.pad(img_test, ((10, 10), (0, 0)), mode='constant', constant_values=255)
print('Image shape =', img_test.shape)
plt.imshow(img_test, cmap='gray')
```

```python
# function to pad the images
def padding(img):
    pad_top = (80 - img.shape[0]) // 2
    pad_bottom = 80 - img.shape[0] - pad_top
    pad_left = (60 - img.shape[1]) // 2
    pad_right = 60 - img.shape[1] - pad_left

    return np.pad(img, ((pad_top, pad_bottom), (pad_left, pad_right)), mode='constant', constant_values=255)
```

```python
# loading the images
images = []  # list that will receive all the images
labels = []  # list that will receive all the labels
images_ID = []  # list that will receive the IDs of all the images
wrong_shape = []  # list with the files with a shape different than 80 x 60 that were not loaded

for item in tqdm(os.listdir(images_path)):  # looping through each file in this path
    if item.endswith('.jpg'):  # working with .jpg image files only
        img = cv2.imread(os.path.join(images_path, item), cv2.IMREAD_GRAYSCALE)  # reading the image in grayscale
        img_name = int(item.split('.')[0])

        # padding the images when necessary
        if img is not None and not img.shape == (80, 60):  # checking if the shape is not 80 x 60
            wrong_shape.append([img_name, img.shape[0], img.shape[1]])  # storing the file with different shape's information
            img = padding(img)  # padding the image

        # normalizing the images and storing them
        if img is not None:
            images.append(img.astype('float32') / 255.0)  # normalizing storing the image in a list
            images_ID.append(img_name)  # storing the image's name in a list
            if img_name in id_original:  # encoding the labels
                labels.append([1])  # original image
            else:
                labels.append([0])  # fake image

# converting the lists into arrays
images = np.array(images)
images_ID = np.array(images_ID)
labels = np.array(labels)
```

```python
# checking the images that had a shape different than 80 x 60 before padding
wrong_shape = pd.DataFrame(wrong_shape)
print(wrong_shape)
```

There were 23 different images with different shapes.

## Splitting the data

The specific rule for this step is: "the test data must include at least 1000 product images".

### Sample size

Defining the sample size that will be used for modelling. We will split it into train, val, and test.

```python
SAMPLE_SIZE = int(0.2 * images.shape[0])
print('The sample size is', SAMPLE_SIZE)
```

Originally, 34.1% of the images were original. We can maintain this proportion in the sample.

```python
# selecting the sample
X_sample, _, y_sample, _ = train_test_split(images, labels, train_size=SAMPLE_SIZE, random_state=42, stratify=labels)
print('Shape of X_sample =', X_sample.shape)
print('Shape of y_sample =', y_sample.shape)
print(f'The proportion is {y_sample.sum() / len(y_sample) * 100:.1f}%')
```

```python
# splitting the sample into training and (validation + test) datasets
X_train, X_val, y_train, y_val = train_test_split(X_sample, y_sample, test_size=0.3, random_state=42, stratify=y_sample)
```

```python
# splitting the (validation + test) dataset into separate validation and test datasets
X_val, X_test, y_val, y_test = train_test_split(X_val, y_val, test_size=0.5, random_state=42, stratify=y_val)
```

```python
# deleting the variables that won't be used anymore
del img, img_name, img_test, item, X_sample, y_sample
```

```python
print('Shape of X_train =', X_train.shape)
print('Shape of y_train =', y_train.shape)
print(f'The proportion of train data is {y_train.sum() / len(y_train) * 100:.1f}%\n')

print('Shape of X_val =', X_val.shape)
print('Shape of y_val =', y_val.shape)
print(f'The proportion of val data is {y_val.sum() / len(y_val) * 100:.1f}%\n')

print('Shape of X_test =', X_test.shape)
print('Shape of y_test =', y_test.shape)
print(f'The proportion of test data is {y_test.sum() / len(y_test) * 100:.1f}%')
```

# CNN Model 1

This model identifies if a product image seems to belong to the original brand or has been uploaded with a fake brand.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Flatten, Dense
```

```python
#Simple Model

model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(80, 60, 1)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
from tensorflow.keras.optimizers.legacy import Adagrad

# Compiling the model with Adagrad optimizer
model.compile(optimizer=Adagrad(), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Training the model
history1 = model.fit(X_train.reshape(-1, 80, 60, 1), y_train, epochs=10, batch_size=32, validation_data=(X_val.reshape(-1, 80, 60, 1), y_val))
```

```python
# Evaluating the model
test_loss_adagrad, test_accuracy_adagrad = model.evaluate(X_test.reshape(-1, 80, 60, 1), y_test)
print(f'Test Accuracy (Adagrad): {test_accuracy_adagrad*100:.2f}%')
```

```python
from tensorflow.keras.optimizers.legacy import Adam

# Compiling the model with Adam optimizer
model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Training the model
history2 = model.fit(X_train.reshape(-1, 80, 60, 1), y_train, epochs=10, batch_size=32, validation_data=(X_val.reshape(-1, 80, 60, 1), y_val))
```

```python
# Evaluating the model
test_loss_adam, test_accuracy_adam = model.evaluate(X_test.reshape(-1, 80, 60, 1), y_test)
print(f'Test Accuracy (Adam): {test_accuracy_adam*100:.2f}%')
```

```python
from tensorflow.keras.layers import MaxPooling2D, Dropout

# Building Complex model
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(80, 60, 1)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))

# Flatten layer
model.add(Flatten())

# Dense layers
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))

# Display the model summary
model.summary()
```

```python
# Compile the model
model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
history1 = model.fit(X_train.reshape(-1, 80, 60, 1), y_train, epochs=10, batch_size=32, validation_data=(X_val.reshape(-1, 80, 60, 1), y_val))
```

```python
# Evaluating the model
test_loss, test_accuracy = model.evaluate(X_test.reshape(-1, 80, 60, 1), y_test, batch_size=32)
print(f'Test Accuracy: {test_accuracy*100:.2f}%')
```
